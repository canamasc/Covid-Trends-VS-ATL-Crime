```python
# Crime data is Feb 2 2020 thru Sep 29 2020
# Covid data is Feb 29 2020 thru July 21 2021

# Let's trim crime data to start Feb 29 2020 
actualstarthead = dfcrime.loc[dfcrime['CMPLNT_FR_DT'] == '2/29/2020']
actualstart = actualstarthead.head(1).iloc[0].name
n = dfcrime['CMPLNT_FR_DT'].count() - actualstart
print(dfcrime['CMPLNT_FR_DT'].count())
print(n)
print(actualstart)
dfcrime = (dfcrime.tail(n)).rename(columns={'CMPLNT_FR_DT' : "DATE"})
dfcrime


```

```python
#trim covid data to end Sep 29 2020
dfcovid = dfcovid.head(dfcovid.loc[(dfcovid['DATE_OF_INTEREST'] == '09/30/2020')].index[0])
dfcovid = dfcovid.rename(columns={'DATE_OF_INTEREST':'DATE'})
dfcovid
```

```python
# dates in slightly diff formats
newdates = []

for index,row in dfcovid.iterrows():
    if row['DATE'][3] == '0':
        row['DATE'] = row['DATE'][:3] + row['DATE'][4:]
    if row['DATE'][0] == '0':
        newdates.append(row['DATE'][1:])
dfcovid['DATE'] = newdates
dfcovid
```

```python
# merge the two dataframes
merged = pd.merge(dfcrime, dfcovid, how='left')
merged
```

```python
# Fill in empty values
for col in merged.columns:
    merged[col] = merged[col].fillna('N/A') 
merged
```

```python
# Export to csv
merged.to_csv('merged_NYC_data.csv')
```
