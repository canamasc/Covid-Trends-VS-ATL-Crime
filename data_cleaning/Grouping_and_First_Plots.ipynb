```python
dfcovid['county'].value_counts()
dfcrime = dfcrime.loc[dfcrime['county'] == 'Fulton']
dfcovid = dfcovid.loc[dfcovid['county'] == 'Fulton']
```

```python
dfcrime
```

```python
output_df = pd.merge(dfcrime, dfcovid, how='left')
output_df
```

```python
#meep2 = df1.groupby(['SN'])['Price'].mean()
#meep1 = df1.groupby(['SN'])['Purchase ID'].count()
#uniqueSNs = uniqueSNs.assign(snPurchaseCounts=meep1.values)

m = ((output_df.groupby(['use_date']).count())['deaths'])#.
output_df = (pd.merge(output_df, m, on= 'use_date',how='left')).rename(columns={'deaths_y': 'daily_reports', 'deaths_x': 'deaths'})
m = (output_df.groupby(['use_date','neighborhood']).count())['lat']
output_df = (pd.merge(output_df, m, on= ['use_date','neighborhood'],how='left')).rename(columns={'lat_y': 'daily_reports_nbhd', 'lat_x': 'lat'})

m = (output_df.groupby(['use_date','UC2_Literal']).count())['lat']
output_df = (pd.merge(output_df, m, on= ['use_date','UC2_Literal'],how='left')).rename(columns={'lat_y': 'daily_reports_crimetype', 'lat_x': 'lat'})

#neighborhood_cases = output_df.groupby(['neighborhood'])['total_cases'].sum()
#neighborhood_cases
```

```python
# For neighborhood = Downtown, plot daily_reports vs time
import scipy.stats as st
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np

x_values= output_df['total_cases']
y_values = output_df['daily_reports']

plt.scatter(x_values, y_values)
plt.title('Crime vs Covid Cases')
plt.xlabel('total covid cases per day')
plt.ylabel('daily crime reports')

# Linear regression model

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_values,regress_values,"r-")
print("Line equation is " + line_eq)

# investigate outlier
plt.show()
```
