```python
import pandas as pd
```

```python
data1 = pd.read_csv('crime_data1.csv')
data2 = pd.read_csv('covid_data.csv')

output_df = pd.merge(data1, data2, how='left')

print(output_df)
```

```python
clean_df = output_df.dropna(how='any', inplace=False)
clean_df.drop(['rpt_date', 'occur_time'], inplace=True, axis=1)
crime_covid_df = pd.DataFrame(clean_df)
#crime_covid_df.head(-1)
crime_covid_df.sort_values(by='use_date')
```
