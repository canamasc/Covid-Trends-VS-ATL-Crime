```python
import pandas as pd
```

```python
covid_df = pd.read_csv('epicurve_rpt_date.csv')

use_df = covid_df[covid_df["county"]!="Georgia"]

use_df
```

```python
use_df['use_date'] = pd.to_datetime(use_df['report_date'])

use1_df = use_df[((use_df["use_date"].dt.year==2020)&((use_df["use_date"].dt.month==9)|(use_df["use_date"].dt.month==10)|(use_df["use_date"].dt.month==11)|(use_df["use_date"].dt.month==12)))|((use_df["use_date"].dt.year==2021)&((use_df["use_date"].dt.month==1)|(use_df["use_date"].dt.month==2)|(use_df["use_date"].dt.month==3)))]
use1_df
```

```python
use1_df = use1_df[["county","use_date","total_cases","deaths"]]

use1_df
```

```python
use1_df = use1_df[(use1_df["county"]=="Fulton")|(use1_df["county"]=="Dekalb")|(use1_df["county"]=="Clayton")]

use1_df
```

```python
final_df = use1_df.dropna(how='any')
```

```python
final_df.to_csv("covid_data.csv", index=False, header=True)
```
